# Проект: телекоммуникации

Оператор связи «ТелеДом» хочет бороться с оттоком клиентов. Для этого его сотрудники начнут предлагать промокоды и специальные условия всем, кто планирует отказаться от услуг связи. Чтобы заранее находить таких пользователей, «ТелеДому» нужна модель, которая будет предсказывать, разорвёт ли абонент договор. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и услугах. Ваша задача — обучить на этих данных модель для прогноза оттока клиентов.

**Описание услуг**

Оператор предоставляет два основных типа услуг: 

  - Стационарную телефонную связь. Телефон можно подключить к нескольким линиям одновременно.
  
  - Интернет. Подключение может быть двух типов: через телефонную линию (DSL, от англ. digital subscriber line — «цифровая абонентская линия») или оптоволоконный кабель (Fiber optic).
  
  
Также доступны такие услуги:

  - Интернет-безопасность: антивирус (DeviceProtection) и блокировка небезопасных сайтов (OnlineSecurity);
  
  - Выделенная линия технической поддержки (TechSupport); 
  
  - Облачное хранилище файлов для резервного копирования данных (OnlineBackup);
  
  - Стриминговое телевидение (StreamingTV) и каталог фильмов (StreamingMovies).
  
  
Клиенты могут платить за услуги каждый месяц или заключить договор на 1–2 года. Возможно оплатить счёт разными способами, а также получить электронный чек.


**Описание данных**

Данные состоят из нескольких файлов, полученных из разных источников:

  - contract_new.csv — информация о договоре;
  
  - personal_new.csv — персональные данные клиента; 
  
  - internet_new.csv — информация об интернет-услугах;
  
  - phone_new.csv — информация об услугах телефонии.
  
***Файл contract_new.csv***

- customerID — идентификатор абонента;

- BeginDate — дата начала действия договора;

- EndDate — дата окончания действия договора;

- Type — тип оплаты: раз в год-два или ежемесячно;

- PaperlessBilling — электронный расчётный лист;

- PaymentMethod — тип платежа;

- MonthlyCharges — расходы за месяц;

- TotalCharges — общие расходы абонента.

***Файл personal_new.csv***
 
- customerID — идентификатор пользователя;

- gender — пол;

- SeniorCitizen — является ли абонент пенсионером;

- Partner — есть ли у абонента супруг или супруга;

- Dependents — есть ли у абонента дети.


***Файл internet_new.csv***
 
- customerID — идентификатор пользователя;

- InternetService — тип подключения;

- OnlineSecurity — блокировка опасных сайтов;

- OnlineBackup — облачное хранилище файлов для резервного копирования данных;

- DeviceProtection — антивирус;

- TechSupport — выделенная линия технической поддержки;

- StreamingTV — стриминговое телевидение;

***Файл phone_new.csv***
 
- customerID — идентификатор пользователя;

- MultipleLines — подключение телефона к нескольким линиям одновременно.

Во всех файлах столбец customerID содержит код клиента. Информация о договорах актуальна на 1 февраля 2020 года.

#### Построим круговые диаграммы для категориальных признаков

```python
# Функция для построения круговых диаграмм
def plot_pie_histograms(df, columns_cat):
    for column in columns_cat:
        if column in df.columns:
            sizes = df[column].value_counts()
            sizes.plot.pie(autopct='%.1f%%', figsize=(6, 6), startangle=90)
            plt.title(f'Круговая диаграмма для "{column}"')
            plt.ylabel('')
            plt.show()
```

Для визуализации распределения этих признаков построим виолончельные графики. Они позволяют наглядно сравнить, как числовой признак распределён в разных группах клиентов. Такие графики показывают не только среднее значение и разброс, но и форму распределения — где значения наиболее сконцентрированы, а где встречаются реже.
```python
def violent_plot(df, columns, target, corr_matrix=None):
    for column in columns:
        print(f"Виолончельный график для столбца: {column}")

        # Виолончельный график
        plt.figure(figsize=(8, 5))
        sns.violinplot(x=target, y=column, data=df, split=True)
        plt.title(f"Violin plot для {column}")
        plt.show()

        if corr_matrix is not None and target in corr_matrix.columns:
            correlation = corr_matrix.loc[column, target]
            print(f"Корреляция Phik для {column}: {round(correlation, 2)}")
        
        print("-" * 70)

# Визуализация и вывод для каждого столбца
violent_plot(df, columns_to_plot, 'target', corr_matrix)
```

**Рассмотрим оставшиеся признаки.**

```python
def plot_custom(data, columns, phik_matrix=None):
    for column in columns:
        records = []

        for value in data[column].unique():
            total_count = data[data[column] == value].shape[0]
            churned = data[(data[column] == value) & (data['target'] == 1)].shape[0]
            not_churned = total_count - churned

            share_churn = np.round(churned / total_count, 2)
            share_stay = np.round(not_churned / total_count, 2)

            records.append({'Значение': value, 'Статус': 'Ушёл', 'Доля': share_churn})
            records.append({'Значение': value, 'Статус': 'Остался', 'Доля': share_stay})

        plot_df = pd.DataFrame(records)

        # Красиво выводим таблицу
        styled_table = plot_df.pivot(index='Значение', columns='Статус', values='Доля')
        display(styled_table.style.format("{:.2f}").background_gradient(axis=1))

        # Строим график
        plt.figure(figsize=(8, 5))
        sns.barplot(data=plot_df, x='Значение', y='Доля', hue='Статус')
        plt.title(f'Распределение по признаку: {column}', fontsize=10)
        plt.ylabel('Доля клиентов')
        plt.xlabel(column)
        plt.xticks(rotation=30)
        plt.tight_layout()
        plt.show()

        # Показываем корреляцию
        if phik_matrix is not None:
            corr_value = np.round(phik_matrix.loc[column]['target'], 2)
            print(f'Phik-корреляция между "{column}" и целевой переменной: {corr_value}')
        print('-' * 75)
```

```python
# Делим данные на тренировочные и тестовые
features_train, features_test, target_train, target_test = train_test_split(features, target,
                                                                            test_size = TEST_SIZE,
                                                                            random_state = RANDOM_STATE)

print(features_train.shape, features_test.shape)
print(target_train.shape, target_test.shape)
```

```python
# Масштабируем числовые данные и кодируем категориальные признаки с использованием OneHotEncoder
# Определяем числовые и категориальные столбцы
num = list(features.select_dtypes(include=['int64', 'float64']).columns)
cat = list(features.select_dtypes(include='object').columns)

# Создаём преобразователь для числовых и категориальных данных
col_transformer = ColumnTransformer(
    transformers=[
        # Масштабируем числовые данные
        ('scaler', StandardScaler(), num), 
        
        # Кодируем категориальные данные
        ('ohe', OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore'), cat)  
    ],
    remainder="passthrough"  # Остальные столбцы оставляем без изменений
)
```

## Обучение моделей машинного обучения

```python
# Создадим датафрейм для вывода результатов
results = pd.DataFrame(columns=['ROC_AUC', 'Precision'])

def results_table(name, roc_auc, precision):
    results.loc[name] = [roc_auc, precision]
    return results
```

### Модель RandomForestClassifier

```python
grid_params = { 
    'n_estimators': [50, 70, 90],
    'max_depth' : [12, 16, 20]
}

grid_rf = GridSearchCV(RandomForestClassifier(class_weight='balanced', random_state = RANDOM_STATE), 
                       grid_params, 
                       cv=3, 
                       scoring=['roc_auc', 'precision'], 
                       refit = 'roc_auc').fit(features_train, target_train)
```

```python
best_rf = grid_rf.best_estimator_

cv_results = pd.DataFrame.from_dict(grid_rf.cv_results_)
best_idx = grid_rf.best_index_
roc = cv_results.loc[best_idx, 'mean_test_roc_auc']
pr = cv_results.loc[best_idx, 'mean_test_precision']

results_table('RandomForestClassifier', round(roc, 4), round(pr, 4))
```

- Кросс-валидация: используется 3-кратная кросс-валидация (cv=3) для оценки качества модели.


- Балансировка классов: включена через параметр class_weight='balanced', что автоматически учитывает дисбаланс в выборке.


- Подбор гиперпараметров:

   - n_estimators: [50, 70, 90] — число деревьев в ансамбле.

### Модель CatBoostClassifier

```python
model_cbr = CatBoostClassifier(
    loss_function='Logloss',
    auto_class_weights='Balanced',
    verbose=0,
    random_state = RANDOM_STATE
)

parameters_cbr = {
    'max_depth': [4, 6, 8],
    'learning_rate': [0.01, 0.1],
    'iterations': [500, 1000]
}

catboost_grid = GridSearchCV(model_cbr,
                             parameters_cbr,
                             cv=3,
                             scoring=['roc_auc', 'precision'],
                             refit = 'roc_auc').fit(features_train, target_train)

cv_results = pd.DataFrame.from_dict(catboost_grid.cv_results_)
best_idx = catboost_grid.best_index_
roc = cv_results.loc[best_idx, 'mean_test_roc_auc']
pr = cv_results.loc[best_idx, 'mean_test_precision']

results_table('CatBoostClassifier', round(roc, 4), round(pr, 4))
```
